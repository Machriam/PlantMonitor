```python
using System;
using Plotly.NET.Interactive;
using Plotly.NET;
```

```python
var rampLength=100;
var x=Enumerable.Range(1,rampLength).Select(x=>(double)x).ToList();
var k=0.12;
var minTime=50;
var maxTime=2000;
var L=maxTime-minTime;
var x0=1*minTime;
Func<double,double> logisticFunction=(x)=>maxTime-L/ (1+Math.Pow(Math.E,(-k*(x-x0))));
var y=x.Select(logisticFunction).ToList();

var scatter=Chart2D.Chart.Scatter<double,double,string>(x:x,y:y,mode:StyleParam.Mode.Lines);


scatter

```
